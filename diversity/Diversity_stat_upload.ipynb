```python
def fig_diversity(x, h, err, cp=3, ylim=(0,0.03), ax=None, color='red', title=None, xtick_lab=None,
                  bottom=False, top = True, xlab=None, ylab=None, fig_name=None, side=None):
  #define ax
  if ax is None:
    fig, ax = plt.subplots()
    sns.despine(ax=ax, bottom=bottom, top = top, offset=5)

  #plot
  ax.bar(x=x,height=h, yerr=err, color=color,capsize=cp,tick_label =None)

   # Plot title
  if title:
    ax.set_title(f'{title}')

  # Customize plot
  if ylab:
    ax.set_ylabel(f'{ylab}')
  if xlab:
    ax.set_xlabel(f'{xlab}')

  #ax setting
  ax.set_ylim(*ylim)
  if xtick_lab:
    if top is True:
      ax.set_xticklabels([xtick_lab[v] for v in ax.get_xticks()], rotation=30, ha='right')
    else:
      ax.set_xticklabels([xtick_lab[v] for v in ax.get_xticks()], rotation=30, ha='left')

  if fig_name:
    fig.savefig(f"drive/MyDrive/Genomic/diversity_stats/diversity_{fig_name}.png", dpi=100, bbox_inches='tight', transparent=False)
```

```python
def compute_accessible_length(region='3L', site_mask='gamb_colu_arab'):
  access = ag3.is_accessible(region=region, site_mask='gamb_colu_arab')
  l = np.count_nonzero(access)
  return l
```

```python
pop_id = ['Bana_gam_2016']
def plot_sfs_year(sf, nf, figsize=(12,12), legend=False, ax=None, lw=2, fn=None, ylab=False, xlab=False):
    l = compute_accessible_length(region='3L', site_mask='gamb_colu_arab')
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)
    # set order for Z
    for pop in pop_id:
      s1 = sf[pop] / l
      n1 = nf[pop]
      allel.plot_sfs_folded_scaled(s=s1, n=n1, label=pop, ax=ax,
                                   plot_kwargs=dict(color=pop_colors_year[pop],lw=lw, alpha=0.8))
    # ax setting
    if ylab == True:
        ax.set_ylabel('SNPs density (pb$^{-1}$)', fontsize=12)
    else:
        ax.set_ylabel('')

    if xlab == True:
        ax.set_xlabel('Minor allelic frequencies (MAF)', fontsize=12)
    else:
        ax.set_xlabel('')
    ax.set_ylim(0, 0.06)
    ax.set_xlim(0, 0.5)

    #Legend
    handles = [mpl.patches.Patch(color=pop_colors_year[pop], label=pop_labels_year[pop]) for pop in pop_id]
    ax.legend(handles=handles, loc='upper right', ncol=1, labelspacing=0.8)
    if fn:
        fig.savefig(fn,  dpi=300, bbox_inches='tight');
```

```python
## An. gambiae
fig, ax = plt.subplots(figsize=(24,18))
gs = GridSpec(6, 4)

ax0 = plt.subplot(gs[0, 0])
pop_id = ['Bana_gam_2016']
plot_sfs_year(sf=sf1, nf=nf1, legend=False, ax=ax0, lw=2, fn=None, ylab=False, xlab=False)
```
