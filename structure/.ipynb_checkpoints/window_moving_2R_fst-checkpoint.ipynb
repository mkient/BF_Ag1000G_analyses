```python
#Select BF pops in the metadata
bf_samples = df_samples.query('country == "Burkina Faso" and year > 2004')

#pop_cohort
coh_sp = dict([(f"{s}", list(df.index)) for (s), df in bf_samples.reset_index().groupby(['aim_species'])])

#remove some populations 
coh_rm=['intermediate_gambiae_coluzzii']
list_coh_sp = list(coh_sp.keys())

# remove pop_function
def remove_pop(list1, list2):
  for item in list2:
    list1.remove(item)
  return list1
coh_sh_r = remove_pop(list_coh_sp.copy(), coh_rm)
```

```python
#To the genomics dataset of bf samples 
bf_snps = chrom2R_gt.sel(samples=df_samples.eval('country=="Burkina Faso" and year > 2004'))

# Take some SNP in the 2R chromosomes 
bf_snps_pos = allel.SortedIndex(bf_snps['variant_position'].values)

# To filter the SNP dataset and warp the dataset to GT array 
filt = 'gamb_colu_arab' 
filt_val = bf_snps[f"variant_filter_pass_{filt}"].values 
gt_filtered = allel.GenotypeDaskArray(bf_snps["call_genotype"][filt_val].data)
```

```python
# compute allele counts
ac_sp = {}
with ProgressBar():
  for key in ['arabiensis', 'coluzzii', 'gambiae']:
    ac_sp[key] = gt_filtered.take(coh_sp[key], axis=1).count_alleles(max_allele=3).compute()
```

# Fst

```python
@functools.lru_cache(maxsize=None)
def load_ac(pop):
    ac_spi = ac_sp[pop]
    return ac_spi
#load_ac(pop='arabiensis')
```

```python
#@functools.lru_cache(maxsize=None)
def compute_windowed_pair_fst(pops, filt_pos, size):
    # order is irrelevant
    pop1, pop2 = pops
    ac_pop1 = load_ac(pop=pop1)
    ac_pop2 = load_ac(pop=pop2)
    loc_pass = ac_pop1.is_segregating() & ac_pop2.is_segregating() & (ac_pop1.max_allele() <= 1) & (ac_pop2.max_allele() <= 1)
    # Find selected SNPs pos
    pos=filt_pos[loc_pass]
    # compute windowed_fst
    fst_wind,windows,counts = allel.windowed_hudson_fst(pos, ac_pop1[loc_pass], ac_pop2[loc_pass],size=size)
    return fst_wind,windows,counts 
  
def compute_windowed_pairwise_fst(coh_list, pos_filt, size):
    pairwise_fst = {}
    wind_pairwise_fst = {}
    for pop1, pop2 in itertools.combinations(coh_list, 2):
        pops = tuple(sorted([pop1, pop2]))
        fst, wind, _ = compute_windowed_pair_fst(pops=pops,filt_pos=pos_filt,size=size)
        pairwise_fst[f'fst_{pop1[:3]}_{pop2[:3]}'] = fst
        wind_pairwise_fst[f'wind_{pop1[:3]}_{pop2[:3]}'] = wind
    return pairwise_fst, wind_pairwise_fst
```

```python
# compute fst and windows
fst_dict, wind_fst_dict = compute_windowed_pairwise_fst(coh_list=coh_sh_r, pos_filt=filtered_pos, size=100000) 

# warp fst data to dataframe
len(fst_dict['fst_ara_col']) == len(fst_dict['fst_ara_gam'])==len(fst_dict['fst_col_gam'])
df_fst=pd.DataFrame(fst_dict)
len(wind_fst_dict['wind_ara_col']) == len(wind_fst_dict['wind_ara_gam']) == len(wind_fst_dict['wind_col_gam'])==True
df_wind=pd.DataFrame(np.mean(wind_fst_dict['wind_ara_col'], axis=1), columns=['wind_mean'])
data_fst=pd.concat([df_wind, df_fst], axis=1)
print('done')
```

```python
# Plotting FST
# Compute the figure of sequence divergence 
def fig_fst(x, y, title=None,  ax=None, fig_name=None):
  #define ax 
  if ax is None:
    fig, ax = plt.subplots(figsize=(12, 4))
    sns.despine(ax=ax, offset=5)

  # make plot 
  ax.plot(x, y, linestyle='-', marker=None, color='darkred', linewidth=0.4)
  # Plot title
  if title:
    ax.set_title(f'{title}')
  
  # Customize plot 
  ax.set_ylabel('Pairwise  $F_{ST}$')
  ax.set_xlabel('Chrom 2R (bp)')
  xlim=(list(x)[0], list(x)[-1])
  ax.set_xlim(*xlim)
  ax.set_xticklabels(['{:,}'.format(int(v)) for v in ax.get_xticks()]) 
  if fig_name:
    fig.savefig(f"drive/MyDrive/Genomic/graph/2R_{fig_name}.png", dpi=300, bbox_inches='tight')
```

```python
fig_fst(x=data_fst.wind_mean,y=data_fst.fst_ara_col, title='Pairwise $F_{ST}$ between An. arabiensis and An. coluzzii',
        fig_name='Fst_ara_col')
fig_fst(x=data_fst.wind_mean,y=data_fst.fst_ara_gam, title='Pairwise $F_{ST}$ between An. arabiensis and An. gambiae',
        fig_name='Fst_ara_gam')
fig_fst(x=data_fst.wind_mean,y=data_fst.fst_col_gam, title='Pairwise $F_{ST}$ between An. coluzzii and An. gambiae',
        fig_name='Fst_col_gam')
```

```python
melt_df_fst = data_fst.melt('wind_mean', var_name='Crossed-Pop', value_name='pair_fst')
xlim1 = list(data_fst.wind_mean)
# Plot sequence divergence per population 

fig, ax = plt.subplots(figsize=(12, 4))
sns.despine(ax=ax, offset=5)
ax = sns.lineplot(x = 'wind_mean', y = 'pair_fst', hue='Crossed-Pop', data=melt_df_fst,   linewidth=.5)
ax.set(title = 'Pairwise $F_{ST}$ between An. gambiae species', xlabel = 'Chrom 2R (bp)', 
       ylabel = 'Pairwise $F_{ST}$ in 100kb windows')
ax.legend(loc="upper right")
xlim=[xlim1[0], xlim1[-1]]
ax.set_xlim(*xlim)
ax.set_xticklabels(['{:,}'.format(int(v)) for v in ax.get_xticks()])
fig.savefig("drive/MyDrive/Genomic/graph/2R_pair_fst.png", dpi=300, bbox_inches='tight')
```
